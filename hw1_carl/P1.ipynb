## My code with explanations

### Find the period
Here I use a scipy function "find_peak_cwt" to find local maximum as peaks.  
Then I define a function to count the peaks and the period is equal to the number of years-converted-units divided by the number of peaks.

```python
from scipy import signal
def find_period(data):
    num_years = x[-1]-x[0]
    num_conv = len(x)
    num_peaks = len(signal.find_peaks_cwt(data, [0.1,0.15])) +1
    p = num_conv / num_peaks
    p_years = num_years / num_peaks
    return p, p_years

p, p_years = find_period(y)
print("the period is ",'%.3f'% p,"or", '%.3f'%p_years,"natural years")
```

I made a plot to show the peaks found is corresponding to actual peaks. Notice the last one is not included so I add that one manualy in the function.

```python
plt.plot( x, y )
k = signal.find_peaks_cwt(y, [0.1 ,0.15])
x_index = x[k]
y_index = y[k]
plt.scatter( x_index, y_index, s=10,c='r')
plt.xlabel("Date")
plt.ylabel('CO${_2}$, ppm')
plt.show()
```

```python
# get FFT  and plot raw results
Y=fft(y)
Y_abs = abs(Y)
n = np.arange(len(Y_abs))


plt.plot(n,Y_abs)
plt.yscale('log')
plt.xlabel("Fourier index")
plt.ylabel("fft Intensity ")
i=len(signal.find_peaks_cwt(Y_abs,[1,5]))
plt.scatter(n[i],Y_abs[i],s=8,c='r')
plt.show()
print("the frequency of peaks is",i)
```

### Converts between the “Fourier index” and actual frequency
The N is the length of the data while T is the actual total time in seconds. 

```python
# -----find the actual frequency-------
def actual_freq(year):
    N = len(y)
    T = (year[-1]-year[0]) * 365 * 24 * 3600
    global step
    step = N // 2
    freqindex = np.linspace(0,N/T,step)/2 
    return freqindex

freqindex = actual_freq(x)

plt.plot(freqindex,Y_abs[:step])
plt.yscale('log')
plt.xlabel("Frequency (Hz)")
plt.ylabel("fft Intensity ")
plt.show()
```
